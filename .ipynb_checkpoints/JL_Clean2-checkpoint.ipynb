```python
"""
CLEANING AND EDA for [non-] non-numeric substitutions
Created on Saturday November 4th, 2023
Author: Erin Swan-Siegel
"""

import pandas as pd

#Correct for character error
# data = pd.read_csv('Kaggle_data.csv')
# data.head()
# data.replace(to_replace='?')

df = pd.read_csv(r"C:\Users\erins\OneDrive\Desktop\MS_Data_analytics\44-688\Capstone_Project\scraping-glassdoor-selenium\Company_Rating_testing.csv", encoding='latin-1')
```

```python
df = df.drop(['Number of Reviews','Average Salary','Job Openings','Location Website', 'Company Webpage', 'Global Size 2', 'Company Location', 'Industry 2', 'Headquarters','Industry'], axis=1)
```

```python
# Check for 'Unknown', -1

#df = df[df.'Company Name' != ['Unknown', '-1']] # none
df = df[df['Global Size'] != '-1'] # 883 additional
df = df[df['Global Size'] != 'Unknown']
#df = df[df['Industry'] != ['Unknown', '-1']] # 746 additional
#df = df[df['Description'] != ('Unknown', '-1')] # none
df = df[df['Company Rating'] != -1] # 144 additional
#df = df[df['Headquarters'] != ['Unknown', '-1']] # 7 additional
df = df[df['Company Ownership Type'] != 'Unknown'] # 27 additional
df = df[df['Company Ownership Type'] != -1] # 27 additional
df = df[df['Year Founded'] != -1] # 1,053 additional
df = df[df['Est Revenue'] != 'Unknown / Non-Applicable'] # 1,430 additional
df = df[df['Est Revenue'] != '-1'] # 1,430 additional
```

```python
df.count()
```

```python
df_col = df[['Company Rating','Global Size','Est Revenue','Company Ownership Type']]
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
for i in df_col.columns:
    col_num = df_col[i].value_counts()
    #df_col.sort_values([i])
    print("Visualization for %s: Total = %d" % (i, len(col_num)))
    #print(col_num)
    chart = sns.barplot(x=col_num.index, y=col_num, data=df)
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    plt.show()
```
